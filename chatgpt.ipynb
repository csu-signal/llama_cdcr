ECB+ Processing chat-gpt3.5

```python
infile = open(DATA_PATH+'/im_map_ecb.pkl','rb')
data = pickle.load(infile)
keys=data.keys()
```

```python
import os
if os.path.isfile(DATA_PATH+'/ecbplus_1word.pickle'):
  with open(DATA_PATH+'/ecbplus_1word.pickle', 'rb') as file:
    answer_dict=pickle.load(file)
else:
  answer_dict=dict()
len(answer_dict)
```

```python
import time

for k_i,key in enumerate(keys):
  if not key in answer_dict.keys():
    sample=data[key]
    if sample['split']=='test':
      # try:
        sen1=sample['sentence_1']
        sen2=sample['sentence_2']
        time.sleep(3)
        completion = openai.ChatCompletion.create(
          model = "gpt-3.5-turbo",
          temperature = 0.8,
          max_tokens = 2000,
          messages = [
            {"role": "system", "content": "You are a coreference annotator and you have to make a decision about two events marked by <m> and </m> tokens. You are given two sentences. Answer in one word if they are talking about the same event: that is, if they are coreferent."},
            {"role": "user", "content": "sentence_1: {}\
        sentence_2: {}".format(sen1,sen2)},

          ]
        )
        # print(sen1)
        # print(sen2)
        print(k_i)
        print(completion.choices[0].message['content'])
        answer_dict[key]=completion.choices[0].message['content']

      # except:
      #   print("Bad gateway or rate limit")
```

```python
len(answer_dict)
```

```python
with open(DATA_PATH+'/ecbplus_1word.pickle', 'wb') as file:
  pickle.dump(answer_dict,file)
```
