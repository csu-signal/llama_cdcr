ECB+ statistics

```python
infile = open(DATA_PATH+'/im_map_ecb.pkl','rb')
data = pickle.load(infile)
keys=data.keys()
```

number of pairs for train,test, dev

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
train_pairs=dict()
test_pairs=dict()
dev_pairs=dict()
train_tokens=[]
test_tokens=[]
dev_tokens=[]
for k_i,key in enumerate(keys):
    sample=data[key]
    tokens=tokenizer.tokenize(sample['im_generated_sample'])
    if sample['split']=='train':
      train_pairs[key]=dict()
      train_pairs[key]['im_generated_sample']=sample['im_generated_sample']
      train_pairs[key]['tokens']=tokens
      train_tokens.extend(tokens)
      # tokens=encoding.encode(sample['sentence_1'])
      # train_pairs[key]['sentence1_tokens']=encoding.decode(tokens1)
      print('train')
    elif sample['split']=='test':
      test_pairs[key]=dict()
      test_pairs[key]['im_generated_sample']=sample['im_generated_sample']
      test_pairs[key]['tokens']=tokens
      test_tokens.extend(tokens)
      print('test')
    if sample['split']=='dev':
      dev_pairs[key]=dict()
      dev_pairs[key]['im_generated_sample']=sample['im_generated_sample']
      dev_pairs[key]['tokens']=tokens
      dev_tokens.extend(tokens)
      print('dev')



train_pairs_num=len(train_pairs)
test_pairs_num=len(test_pairs)
dev_pairs_num=len(dev_pairs)
print('train pairs num: '+str(train_pairs_num))
print('test pairs num: '+str(test_pairs_num))
print('dev pairs num: '+str(dev_pairs_num))
```

number of tokens for train,test, dev

```python
print('number of train tokens: ',str(len(train_tokens)))
print('number of test tokens: ',str(len(test_tokens)))
print('number of dev tokens: ',str(len(dev_tokens)))
```

unique tokens for train,test, dev

```python
print('number of train unique tokens: ',str(len(set(train_tokens))))
print('number of test unique tokens: ',str(len(set(test_tokens))))
print('number of dev unique tokens: ',str(len(set(dev_tokens))))
```

token length for train,test, dev

```python
import numpy
train_tokens_len=[len(token) for token in train_tokens]
test_tokens_len=[len(token) for token in test_tokens]
dev_tokens_len=[len(token) for token in dev_tokens]
train_len=numpy.average(train_tokens_len)
test_len=numpy.average(test_tokens_len)
dev_len=numpy.average(dev_tokens_len)
print('average length of train tokens: ',str(train_len))
print('average length of test tokens: ',str(test_len))
print('average length of dev tokens: ',str(dev_len))
```

save results

```python
all_pairs=dict()
all_pairs['train_samples']=train_pairs
all_pairs['train_tokens']=train_tokens
all_pairs['train_len']=train_len
all_pairs['test_samples']=test_pairs
all_pairs['test_tokens']=test_tokens
all_pairs['test_len']=test_len
all_pairs['dev_samples']=dev_pairs
all_pairs['dev_tokens']=dev_tokens
all_pairs['dev_len']=dev_len
with open(DATA_PATH+'/ecb_samples.pickle', 'wb') as file:
    pickle.dump(all_pairs,file)
```
